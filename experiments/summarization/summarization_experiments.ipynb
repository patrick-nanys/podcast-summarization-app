### Summarization helpers

```python
import nltk

nltk.download('punkt')
sent_detector = nltk.data.load('tokenizers/punkt/english.pickle')

def split_to_sentences(text: str):
  return sent_detector.tokenize(text)
```

```python
import torch

def batch_tokenize_sentences(sentences: list[str], tokenizer):
  current_sent_idx = 0
  batch_size = tokenizer.model_max_length
  batches: list[torch.Tensor] = []
  for sent in sentences:
    tokenized_sent = tokenizer(sent, return_tensors='pt').input_ids.to(device)
    if len(batches) == 0 or batches[-1].size()[1] + tokenized_sent.size()[1] > batch_size:
      batches.append(tokenized_sent)
    else:
      batches[-1] = torch.cat((batches[-1], tokenized_sent), dim=1)

  return batches
```

```python
def summarize_batches(model, tokenizer, batches: list[torch.Tensor]) -> str:
  summary = ""
  for batch in batches:
    batch_output_ids = model.generate(batch, min_length=100, max_length=200)
    batch_summary = tokenizer.decode(batch_output_ids[0], skip_special_tokens=True)
    summary += batch_summary + '\n-----\n'

  return summary
```

```python
def summarize_text(model, tokenizer, text):
  token_batches = batch_tokenize_sentences(split_to_sentences(text), tokenizer)
  print(f'Split text into {len(token_batches)} part(s).')
  print('Summarizing text...')
  return summarize_batches(model, tokenizer, token_batches)
```

```python
def evaluate_model(model, tokenizer, model_name: str) -> pd.DataFrame:
  print('Starting to evaluate model: ', model_name)
  results = {'model_name': model_name}
  texts = {
      'sample_text': sample_text,
      'long_text': long_text,
      'long_long_text': long_long_text
  }
  for name, text in texts.items():
    print(f'Evaluating model on {name}.')
    results[name + '_summary'] = summarize_text(model, tokenizer, text)

  return pd.DataFrame(results, index=[0])
```
